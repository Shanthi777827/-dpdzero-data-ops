Checking for Duplicates

```python
Agent_roaster = Agent_roaster.dropna(subset=["agent_id","org_id"])
```

```python
dis_sum = dis_sum.dropna(subset=["agent_id","org_id","call_date"])
```

```python
call_logs = call_logs.dropna(subset=["agent_id","org_id","call_date"])
```

Joining DataFrames

```python
call_logs_merged = pd.merge(call_logs,Agent_roaster , on=["agent_id","org_id"], how = "left")
call_logs_merged.head()  #how='left' All rows from the left DataFrame, and matched rows from the right
```

```python
full_data = pd.merge(call_logs_merged,dis_sum,on=["agent_id","org_id","call_date"], how = "left")
```

```python
full_data.head()
```

3.feature Engineering

```python
# connect Rate =complete calls/Total calls
```

```python
full_data["is_completed"] = full_data["status"].str.lower() == "completed"
```

```python
full_data["is_completed"]
```

```python
full_data.columns
```

```python
Agent_Summary = full_data.groupby(["agent_id","call_date"]).agg(Total_Calls = ("call_id","count"),
                                               Unique_Loans = ("installment_id","nunique"),
                                               completed_calls = ("is_completed","sum"),
                                               Avgcall_Duration = ("duration","mean"),
                                               Presence = ("login_time",lambda x: 1 if (x !="Unknown").any() else 0 )
                  ).reset_index()
```

```python
Agent_Summary
```

```python
Agent_Summary["connect_rate"] = (Agent_Summary["completed_calls"] / Agent_Summary["Total_Calls"]).round(2)
```

```python
Agent_Summary["connect_rate"]
```

Merge Agent Name

```python
Agent_Summary= Agent_Summary.merge(Agent_roaster, on="agent_id", how="left")
Agent_Summary["Agent_Name"] =  Agent_Summary["users_first_name"]+ " " + Agent_Summary["users_last_name"]
```

```python
Agent_Summary["Agent_Name"]
```

```python
Agent_Summary.to_csv("Agent_Performance_Summary.csv",index= False)
```
